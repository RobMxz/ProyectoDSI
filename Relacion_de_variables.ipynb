## Relación entre Promedio y Tiempo de viaje

```python
sb.scatterplot(x='traveltime', y='Promedio', data=dataset1, hue='Promedio')
```

Los que viven más lejos tienen tendencia a promedios más bajos (a partir de 3 horas)

## Relación entre Promedio y Repitencias

```python
sb.scatterplot(x='failures', y='Promedio', data=dataset1, hue='Promedio')
```

A mayor número de repitencias, el promedio disminuye (tope = 9)

## Relación ente Promedio y Soporte académico (profesor extra)

```python
sb.scatterplot(x='schoolsup', y='Promedio', data=dataset1, hue='Promedio')
```

Los alumnos que cuenta con un profesor que los apoye, tienen promedios más bajos que los que no los tienen

## Relación entre Promedio y Frecuencia de salidas

```python
sb.scatterplot(x='goout', y='Promedio', data=dataset1, hue='Promedio')
```

En promedio, los estudiantes que salen entre 2 y 3 horas tienen mejores promedios que los demás (menos o más)
