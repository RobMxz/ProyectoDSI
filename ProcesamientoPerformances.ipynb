```python
import pandas as pd

datos = pd.read_csv("student_data.csv")
datos = datos.drop(["G1","G2","G3","school","reason"],axis = 1)
```

# Limpieza de datos


### Eliminación de registros vacios

```python
datos_na = datos.dropna()
```

### Eliminacion de registros duplicados

```python
datos_nodup = datos_na.drop_duplicates()
```

### Revision  de datos limpios

```python
datos_nodup.count()
```

### Variable target: Promedio

# Transformación de categóricas a dummies

### Transformación de categoricas binarias

```python
#M ES 0 Y F ES 1
datos_nodup["sex"] = datos_nodup["sex"].replace("M", 0).replace("F", 1)
```

```python
#URBAN ES 0 Y RURAL ES 1
datos_nodup["address"] = datos_nodup["address"].replace("U", 0).replace("R", 1)
```

```python
#LE ES 0 Y GT3 ES 1
datos_nodup["famsize"] = datos_nodup["famsize"].replace("LE3", 0).replace("GT3", 1)
```

```python
#TOGETHER ES 0 Y APART ES 1
datos_nodup["Pstatus"] = datos_nodup["Pstatus"].replace("T", 0).replace("A", 1)
```

```python
#YES ES 1 Y NO ES 0
datos_nodup["schoolsup"] = datos_nodup["schoolsup"].replace("yes", 1).replace("no", 0)
```

```python
#YES ES 1 Y NO ES 0
datos_nodup["famsup"] = datos_nodup["famsup"].replace("yes", 1).replace("no", 0)
```

```python
#YES ES 1 Y NO ES 0
datos_nodup["paid"] = datos_nodup["paid"].replace("yes", 1).replace("no", 0)
```

```python
#YES ES 1 Y NO ES O
datos_nodup["activities"] = datos_nodup["activities"].replace("yes", 1).replace("no", 0)
```

```python
#YES ES 1 Y NO ES 0
datos_nodup["nursery"] = datos_nodup["nursery"].replace("yes", 1).replace("no", 0)
```

```python
#YES ES 1 Y NO ES 0
datos_nodup["higher"] = datos_nodup["higher"].replace("yes", 1).replace("no", 0)
```

```python
#YES ES 1 Y NO ES 0
datos_nodup["internet"] = datos_nodup["internet"].replace("yes", 1).replace("no", 0)
```

```python
#YES ES 1 Y NO ES 0
datos_nodup["romantic"] = datos_nodup["romantic"].replace("yes", 1).replace("no", 0)
```

```python
datos_nodup.head()
```

### Transformaciones a las otras variables categoricas

```python
datos_nodup["Mjob"] = datos_nodup["Mjob"].replace("other","otherMomJob").replace("health","nurse")
```

```python
datos_nodup["Fjob"] = datos_nodup["Fjob"].replace("other","otherFatherJob")
```

```python
datos_nodup["guardian"] = datos_nodup["guardian"].replace("other","otherGuardian")
```

```python
dummies = pd.get_dummies(datos_nodup["Mjob"], prefix="M_", dtype=int)
datos_nodup = datos_nodup.join(dummies)
```

```python
datos_nodup = datos_nodup.drop(["Mjob"], axis=1)
```

```python
datos_nodup["Fjob"].value_counts()
```

```python
dummies = pd.get_dummies(datos_nodup["Fjob"], prefix="F_", dtype=int)
datos_nodup = datos_nodup.join(dummies)
```

```python
datos_nodup = datos_nodup.drop(["Fjob"], axis=1)
```

```python
datos_nodup["guardian"].value_counts()
```

```python
dummies = pd.get_dummies(datos_nodup["guardian"], prefix="G_", dtype=int)
datos_nodup = datos_nodup.join(dummies)
```

```python
datos_nodup = datos_nodup.drop(["guardian"], axis=1)
```
